## Retrieval-Augmented Generation (RAG)

* Enhances LLMs by giving them access to external, up-to-date knowledge, making responses more factual and relevant without needing constant retraining. 

* Retrieves relevant documents from a database and adding them to the LLM's prompt

* Need fo effective retrieval techniques to give the LLM highly relevant context to generate answer.

* Effective evaluation framework needed to iterate and improve RAG system during initial development and during post-deployment maintenance.

* This course covers - sentence window retrieval and auto-merging retrieval

* Metrics to evaluate LLM question-answering system: context relevance, groundedness, and answer relevance.


### Sentence Window Retrieval
gives an LLM better context by retrieving window of sentences before and after it in the document.

## Auto-merging Retrieval
organizes the document into a tree-like structure where each parent node's text is divided among its child nodes.

## Evaluation - RAG Triad

Context Relevance, Groundedness ( or Faithfulness), Answer Relevance
